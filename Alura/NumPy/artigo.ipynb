#### Estruturas bsicas do NumPy



Escalar: um elemento nico, adimensional. Pode ser um inteiro, float, hexadecimal, caractere, e vrios outros tipos de dado.

```python
import numpy as np
a = np.array(42)
print(a)
```

Array unidimensional: uma lista de escalares onde podemos identificar cada um deles pela sua posio, ou ndice, na lista.  preciso ressaltar que todos os elementos escalares aqui possuem o mesmo tipo, e que se no for especificado durante a definio do array, o Numpy trabalhar uma rotina para determinar o tipo que garanta a caracterstica homognea.

```python
b = np.array([4,5,22,20])
print(b)
```

Array bidimensional: uma lista de arrays unidimensionais, com o formato de uma matriz (tabela), onde precisamos especificar uma posio de linha e uma posio de coluna para localizar um elemento escalar.

```python
c = np.array([[3, 2, 7],
          [4, 9, 1],
          [5, 6, 8]])
print(c)
```

#### Broadcasting

O broadcasting  uma funcionalidade do Numpy que permite trabalhar com operaes em arrays de formas e tamanhos diferentes, sem que seja necessrio criar cpias dos dados. Por esse motivo,  uma tcnica que economiza tempo e memria. Em muitos casos, o broadcasting pode reduzir o nmero de linhas de cdigo necessrias para realizar uma determinada operao.

Por exemplo, na operao abaixo:

```python
d = np.array([[1, 2], [3, 4]])
e = np.array([10, 20])

f = d + e

print(f)
```
