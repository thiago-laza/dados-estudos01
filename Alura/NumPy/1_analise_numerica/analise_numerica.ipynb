---
- multiplicacao

coeficiente angular



```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
a = ((n*np.sum(X*Y) - np.sum(X)*np.sum(Y)))/(n*np.sum(X**2) - (np.sum(X))**2)
```

coeficiente linear



```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X + b
```

```python
np.linalg.norm(Moscow - y)
```
