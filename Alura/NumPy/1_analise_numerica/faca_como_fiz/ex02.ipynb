```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',',usecols=np.arange(1,6,1),skiprows=1)
```

```python
dado
```

```python
dado.shape
```

```python
diametro_laranja = dado[:5000,0]
```

```python
diametro_laranja
```

```python
peso_laranja = dado[:5000,1]
```

```python
peso_laranja
```

```python
diametro_toranja = dado[5000:10000,0]
```

```python
diametro_toranja
```

```python
peso_toranja = dado[5000:10000,1]
```

```python
peso_toranja
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(diametro_laranja,peso_laranja)
plt.legend(['diametro_laranja x peso_laranja'])
```

```python
plt.plot(diametro_toranja,peso_toranja)
plt.legend(['diametro_toranja x peso_toranja'])
```

```python
plt.plot(diametro_laranja,peso_laranja)
plt.plot(diametro_toranja,peso_toranja)
plt.legend(['laranja','toranja'])
```
