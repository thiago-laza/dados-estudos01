laranja

peso x diametro

```python
x = diametro_laranja
y = peso_laranja
n = np.size(x)
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
```

```python
a
```

```python
b = np.mean(y) - a*np.mean(x)
```

```python
b
```

```python
y = a*x + b
```

```python
np.linalg.norm(x - y)
```

```python
plt.plot(peso_laranja,diametro_laranja)
plt.plot(x,y)
```
